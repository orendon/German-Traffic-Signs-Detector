## Class distribution for both TRAIN and TEST samples

The `download` task took care of splitting the dataset 80/20 for each class.

Therefore, some classes lack a representative amount of samples, so data augmentation might help in such cases

```python
def class_distribution(data_dir):
    class_count = [(d, len(os.listdir(os.path.join(data_dir, d)))) for d in os.listdir(data_dir) 
                                                                   if os.path.isdir(os.path.join(data_dir, d))]

    class_count = sorted(class_count, key=lambda x: x[0])
    x_class = [cls for cls, val in class_count]
    y_count = [val for cls, val in class_count]

    plt.figure(figsize=(20, 5))
    plt.title('Images per class on ' + data_dir)
    plt.bar(x_class, y_count)
```

```python
class_distribution('../images/train')
```

```python
class_distribution('../images/test')
```
